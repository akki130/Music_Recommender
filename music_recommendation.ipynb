```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt
```

```python
musicdt=pd.read_csv("LastFM_Matrix.csv")
musicdt
```

```python
musicdt[musicdt["abba"]==1]["user"].count()
```

```python
musicdt['user'].nunique()
```

```python
mdt=musicdt.drop(['user'],axis=1)
mdt
```

```python
mdt.shape
```

```python
from sklearn.metrics.pairwise import cosine_similarity
data_similarity = cosine_similarity(mdt.T)
data_similarity
```

```python
data_similarity.shape
```

```python
mdt1=pd.DataFrame(data_similarity,columns=mdt.columns,index=mdt.columns)
mdt1
```

```python
mdt1.index.nunique()
```

```python
from sklearn.neighbors import NearestNeighbors
m_model=NearestNeighbors(n_neighbors=285)
m_model.fit(mdt1)
```

```python
indices=m_model.kneighbors(mdt1,return_distance=False)
indices
```

```python
mdt2=pd.DataFrame(indices)
mdt2
```

```python
index_array = mdt1.index.to_numpy()
```

```python
mdt2.to_numpy()
```

```python
mdt1.columns[0]
mdt1.index.to_numpy
```

```python
final_model = pd.DataFrame(index_array[mdt2.to_numpy()], index=mdt1.index)
final_model.head()
```

```python
top10=final_model[list(final_model.columns[:11])]
top10
```

```python
top10.to_csv("top10.csv",index_label="Index")
```
